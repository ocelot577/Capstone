Imports

```python
import praw
import json
import time
import pandas as pd
```

Load in credentials / initiate PRAW

```python
json_file = open('Reddit_creds.json')
json_str = json_file.read()
json_data = json.loads(json_str)
```

```python
reddit = praw.Reddit(user_agent=json_data['user_agent'],
                     client_id=json_data['client_id'], client_secret=json_data['client_secret'])
```

Define subreddits to scrape

```python
yang_sub = reddit.subreddit('YangForPresidentHQ')
```

```python
pete_sub = reddit.subreddit('Pete_Buttigieg')
```

```python
sanders_sub = reddit.subreddit('SandersForPresident')
```

```python
harris_sub = reddit.subreddit('Kamala')
```

```python
biden_sub = reddit.subreddit('JoeBiden')
```

```python
subs = [yang_sub,pete_sub,sanders_sub,harris_sub,biden_sub]
```

Create a list of users from recent comments on these subs

```python
users = []
```

```python
#For each sub above, find the 1,000 most recent comments and append the comment author onto the users list
#convert the list to a set to remove duplicates and then back to a list
start = time.time()
for sub in subs:
    for comment in sub.comments(limit=1000):
        users.append(comment.author)
        
print(len(users))
users = list(set(users))
print(len(users))
end = time.time()
print(end-start)
```

Actual scraping code

```python
# For each user in users, scrape their most recent 1,000 comments
# First scrape run the evening of 8/6/19 - 8/7/19
# Second scrape run 8/13/19
start = time.time()
text = []
author = []
subreddit = []
postid = []
counter = 0
for user in users:
    if counter % 10 == 0:
        prog = time.time()
        print(counter)
        print((prog-start)/60)
    try:  #handle cases where redditor is deleted or something
        for comment in user.comments.new(limit=1000):
            text.append(comment.body)
            author.append(comment.author)
            postid.append(comment.id)
            subreddit.append(comment.subreddit.display_name)
    except:
        pass
    counter += 1
end = time.time()
print(end-start)
```

```python
#save scrape data as a dataframe
df = pd.DataFrame([text,author,subreddit,postid],\
                        index=['Comment Text','Author','Subreddit','Post ID']).T
```
